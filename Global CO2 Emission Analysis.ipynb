# Analysis of Global Emissions Data

#### In this Notebook I will use CO2 data to visulize historical pollution of CO2 emission. The dataset contains yearly CO2 emission per country from 1850.

#### The structure is as follows: 
- Preparation of the data
- Analysis of Sweden 
- Analysis of the USA
- Comparison between Sweden and the USA
- CO2 Emissions in 2019
- Global CO2 emission per year
- Comparison between Sweden and China


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('co2-data.csv')

df.head()
```

### Analysis of Sweden

```python
df_swe = df[df['country']=='Sweden']

df_swe.tail()
```

```python
df_swe.set_index('year')
```

```python
# drop rows before 1950
df_swe = df_swe[df_swe.year > 1949]
```

```python
plt.figure(figsize=(12,6))
plt.title('Historical CO2 Emission per capita in Sweden')
plt.plot(df_swe['year'], df_swe['co2_per_capita'], color='blue', label = 'Sweden')
plt.legend()
plt.show()
```

```python
df_swe.info()
```

```python
plt.figure(figsize=(16,6))
plt.title('Historical CO2 Emission per capita in Sweden')
plt.plot(df_swe['year'], df_swe['oil_co2'], color='blue', label = 'Oil')
plt.plot(df_swe['year'], df_swe['gas_co2'], color='green', label = 'Gas')
plt.plot(df_swe['year'], df_swe['coal_co2'], color='red', label = 'Coal')
plt.plot(df_swe['year'], df_swe['flaring_co2'], color='orange', label = 'Flaring')
plt.plot(df_swe['year'], df_swe['cement_co2'], color='purple', label = 'Cement')
plt.legend()
plt.show()
```

### Analysis of United States of America

```python
# drop rows before 1950
df_us = df_us[df_us.year > 1949]
df_us.head()
```

### Amount of CO2 Emissions in 2019

```python
df_2019 = df[df['year']==2019]
df_2019.head()
```

```python
df_2019.info()
```

##### The upcoming section is only including Emission during 2019 for the five main continents 

```python
#Filtering continents on ISO code with value NaN
df_cont = df_2019[df_2019["iso_code"].isnull()]
df_cont.head(100
            )
```

```python
#Only looking at the five main continents
continents = df_cont.loc[(df_cont['country']== 'Europe') | (df_cont['country']== 'Asia') | (df_cont['country']== 'North America')
           | (df_cont['country']== 'Oceania') | (df_cont['country']== 'South America')]
continents
```

```python
continents = continents.dropna(axis='columns')
continents
```

```python
plot_objects1 = plt.subplots(nrows=2, ncols=2, figsize=(16,12))

fig, ((cont_tot, cont_change), (co_sh, cont_abs)) = plot_objects1

barWidth = 0.5

cont_tot.bar(continents['country'], continents['co2_per_capita'], width=barWidth, color='#66CDAA')
cont_tot.set_title('CO2 Emissions per capita per continent 2019')

cont_change.bar(continents['country'], continents['co2_growth_prct'], width=barWidth, color='#458B74')
cont_change.set_title('CO2 growth in percent 2019')

co_sh.bar(continents['country'], continents['share_global_cumulative_cement_co2'], width=barWidth, color='#FF6347', label='Centiment')
co_sh.bar(continents['country'], continents['share_global_cumulative_coal_co2'], width=barWidth, color='#8B3626', bottom=continents['share_global_cumulative_cement_co2'], label='Coal')
co_sh.bar(continents['country'], continents['share_global_cumulative_flaring_co2'], width=barWidth, color='#FF8247', bottom=continents['share_global_cumulative_coal_co2'], label='Flaring')
co_sh.bar(continents['country'], continents['share_global_cumulative_gas_co2'], width=barWidth, color='#E3A869', bottom=continents['share_global_cumulative_flaring_co2'], label='Gas')
co_sh.bar(continents['country'], continents['share_global_cumulative_oil_co2'], width=barWidth, color='#CD8162', bottom=continents['share_global_cumulative_gas_co2'], label='Oil')
co_sh.set_title('CO2 Share Global Cumulative')
co_sh.legend()

cont_abs.bar(continents['country'], continents['co2_growth_abs'], width=barWidth, color='#96CDCD')
cont_abs.set_title('CO2 growth in absolute numbers')

plot_objects1
```

The graphs above shows information about the total CO2 emission, the change in CO2 emission (both in percent and in absolute numbers) and the source of emissions during 2019. In 2019, the Asia has the second lowest CO2 per capita. While other continents are decreasing the CO2 Emissions, the Emssions in Asia is increasing significally. The absolute increase for Asia is larger than the sum of the absoulte decrease of the other continents, which means that the global amount of CO2 emissions increased in 2019.

### Global CO2 Emission per year

```python
df.head()
```

```python
#Only looking at years after 1950
df_group = df[df.year > 1949]

df_group.info()
```

```python
#Only including countries
df_group = df_group[df_group['iso_code'].notna()]

df_group.head()
```

```python
df_group['year'].unique()
```

```python
#Creating a pivot table with index year, column CO2 and the sum of the CO2 emission
df_piv = df_group.pivot_table(index='year', values='co2', aggfunc='sum', margins=False)
df_piv
```

```python
df_piv_gdp = df_group.pivot_table(index='year', values='gdp', aggfunc='sum', margins=False)
df_piv_gdp
```

```python
plt.figure(figsize=(16,6))


plt.subplot(1,2,1)
plt.title('Total Global Historical Co2 Emission ')
plt.plot(df_piv, ':g')

plt.subplot(1,2,2)
plt.title('Total Global Historical GDP')
plt.plot(df_piv_gdp, ':b')
```

The first graph shows the Total Global Historical CO2 Emission. The second graph shows the total global historical GDP and we clearly see that there is an seasonal error in the data. The next section will take a closer look at what is causing the error. 

```python
df1 = df_group[['country', 'year', 'gdp']]
```

```python
df1[(df1['year']==2000) | (df1['year']==2019)]
```

The table shows that many countries does not have a value on years that are not 1980,1990, 2000, 2010. This is the reason to the peaks in the graph. 

### Comparison between the emission in China and Sweden

```python
df_group[(df_group['country']=='Sweden') | (df_group['country']=='Germany') | (df_group['country']=='China') | (df_group['country']=='India')]
```

```python
df_swe1 = df_group[df_group['country']=='Sweden']
df_china = df_group[df_group['country']=='China']
```

```python
df_swe1 = df_swe1[['country', 'year', 'co2_per_capita', 'consumption_co2_per_capita']]
df_china = df_china[['country', 'year', 'co2_per_capita', 'consumption_co2_per_capita']]
```

```python
fig2, comp = plt.subplots(figsize=(14,6))
```

```python
comp.plot(df_swe1['year'], df_swe1['co2_per_capita'], 'b')
comp.plot(df_swe1['year'], df_swe1['consumption_co2_per_capita'], 'b--')
comp.plot(df_china['year'], df_china['co2_per_capita'], 'g')
comp.plot(df_china['year'], df_china['consumption_co2_per_capita'], 'g--')

comp.set_xlabel('Year')
comp.set_ylabel('t ton CO2')

comp.set_title('Comparison between Sweden and China')

comp.legend(['Swe, co2 per capita', 'Swe, consumed co2 per capita', 'China, co2 per capita', 'China, consumed co2 per capita'])

fig2
```

The graph shows the historical emission of CO2 data in Sweden and China, both actual CO2 emission and the consumed CO2 per capita. The blue lines shows that Sweden has for many years emitted a large amount of CO2 emissions per capita, much more than China. China shows an increasing trend, while the trend in Sweden in decreasing since 1970s. The breaking point between Sweden and Chinas emission per capita happened around 2008. The breaking point in CO2 consumed per capita, has not yet happened.

The next graph will show the absolute total amount of CO2 emission that the both countries has emitted since 1950, per capita. Let's also add USA to this comparison.

```python
#Sweden
swe_co2 = df_swe1['co2_per_capita'].sum().round()
swe_cons = df_swe1['consumption_co2_per_capita'].sum().round()
print(swe_co2)
print(swe_cons)
```

```python
#China
chi_co2 = df_china['co2_per_capita'].sum().round()
chi_cons = df_china['consumption_co2_per_capita'].sum().round()
print(chi_co2)
print(chi_cons)
```

```python
#USA
us_co2 = df_us['co2_per_capita'].sum().round()
us_cons = df_us['consumption_co2_per_capita'].sum().round()
print(us_co2)
print(us_cons)
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1,2,1)
plt.title('Sum of total CO2 Emission per capita and year since 1950 ')
plt.bar('Sweden', swe_co2, width=0.5, align='center', color='b')
plt.bar('China', chi_co2, width=0.5, align='center', color='g')
plt.bar('USA', us_co2, width=0.5, align='center', color='r')

plt.subplot(1,2,2)
plt.title('Sum of total consumed CO2 Emission per capita and year since 1990')
plt.bar('Sweden', swe_cons, width=0.5, align='center', color='b')
plt.bar('China', chi_cons, width=0.5, align='center', color='g')
plt.bar('USA', us_cons, width=0.5, align='center', color='r')
```

The graph clearly shows big differences in aggregated CO2 per capita between the three countries.
